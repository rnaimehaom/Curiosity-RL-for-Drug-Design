## Visualize the property shift: G VS G-RL

```python
import pandas as pd

df = pd.read_csv('gen_smis/G_all_smis_.txt', header=None)
G_smis = list(df[0])
len(G_smis)
```

```python
# df = pd.read_csv('gen_smis/RL_G_0_single_best__all_smis_.txt', header=None)
# df = pd.read_csv('gen_smis/RL_G_1_single_best__all_smis_.txt', header=None)
# df = pd.read_csv('gen_smis/RL_G_2_single_best__all_smis_.txt', header=None)
# df = pd.read_csv('gen_smis/RL_G_3_single_best__all_smis_.txt', header=None)
df = pd.read_csv('gen_smis/RL_G_4_single_best__all_smis_.txt', header=None)

biased_smis = list(df[0])
len(biased_smis)
```

```python
from utils import check_validity

val_smis1 = []
for s1 in G_smis:
    v1 = check_validity(s1)
    if v1 == 1:
        val_smis1.append(s1)
        
val_smis2 = []
for s2 in biased_smis:
    v2 = check_validity(s2)
    if v2 == 1:
        val_smis2.append(s2)
```

```python
unq1 = list(set(val_smis1))
unq2 = list(set(val_smis2))
```

```python
from utils import get_SAS

sas1 = []
for s1 in unq1:
    sas = get_SAS(s1)
    sas1.append(sas)
    
sas2 = []
for s2 in unq2:
    sas = get_SAS(s2)
    sas2.append(sas)
```

```python
print(len(sas1), len(sas2))
```

## ----------------------------------------------------------------------

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(x=sas1, shade=True, color="r", label='G')
sns.kdeplot(x=sas2, shade=True, color="b", label='RL-G-0')
plt.legend()

plt.savefig('figs_umaps_kdes/SAS_G_VS_RL_G_0_single.png', dpi=300)
```
